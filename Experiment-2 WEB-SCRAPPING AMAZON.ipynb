```python
url="https://www.amazon.in/gp/aw/reviews/B09DMRPV51/ref=cm_cr_dp_mb_top?ie=UTF8"
```

```python
import requests           ####load url
from bs4 import BeautifulSoup    ####for web  ####to excute instruction press shift+enter button
```

```python
page=requests.get(url)
page
```

```python
page.content                              ####to get html codeof website
```

```python
soup=BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
names=soup.find_all('span',class_="a-profile-name")[2:]      
names
```

```python
len(names)
```

```python
cust_name=[]
for i in range(0,len(names)):cust_name.append(names[i].get_text())
cust_name
```

```python
stars=soup.find_all('span',class_="a-icon-alt")[3:]
stars
```

```python
len(stars)
```

```python
cust_rating=[]
for i in range(0,len(stars)):cust_rating.append(stars[i].get_text())
cust_rating
```

```python
title=soup.select("a.review-title span")
title
```

```python
len(title)
```

```python
review_title=[]
for i in range(0,len(title)):review_title.append(title[i].get_text())
review_title
```

```python
dates=soup.select("span.review-date")[2:]
dates
```

```python
review_date=[]
for i in range(0,len(dates)):review_date.append(dates[i].get_text())
review_date
```
