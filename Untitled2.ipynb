```python
#2.BAR Graph ##
import matplotlib.pyplot as plt
a=[1,2,4,5,6]
b=[6,8,9,10,11]
plt.title("Bar Graph Visualization.")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.bar(a,b,color="blue",label="BAR-GRAPH")
plt.legend(loc=2)
plt.bar(a,b)
plt.show()
```

```python
#Scatter Graph
import matplotlib.pyplot as plt
a=[2,4,5,6,7]
b=[7,8,9,10,11]
plt.title("Scatter Graph Vsualization by Roll No 35")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.scatter(a,b,color="blue",label="scatter plot",s=250,marker="*")
plt.legend(loc=2)
plt.scatter(a,b)
plt.show()
```

```python
#pie Diagram
import matplotlib.pyplot as plt
slices=[30,60,90,120]
plt.title("Pie Graph Vsualization.")
activities=["sleeping","eating","gaming","studying"]
cols=["red","green","blue","yellow"]
plt.pie(slices,labels=activities,colors=cols,autopct="%1.2f%%",
explode=(0.3,0.3,0.3,0.3))
```
